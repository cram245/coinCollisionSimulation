## PDEs: Solution of the 2D Heat Equation using Finite Differences
 This is an example of the **numerical solution** of a **Partial Differential 
 Equation** using the **Finite Difference Method**. 
 
 In this case applied to the **Heat equation**
 $$u_t=k_c (u_{xx}+u_{yy})$$
 To approximate the derivative of a function in a point, we use the finite 
 difference schemes.

#### 1D Finite Differences 
 One can choose different schemes depending on the final wanted precission.
 
 
 
 **Forward Differences**: error $O(h)$
 
$$f'(x_0) \approx \frac{f(x_0+h)-f(x_0)}{h}$$
 
**Central Differences**: error $O(h^2)$
 
$$f'(x_0) \approx \frac{f(x_0+h)-f(x_0-h)}{2h}$$
 
 **Second derivative**. 
 
 **Central Differences**: error $O(h^2)$

$$f''(x_0) \approx \frac{f(x_0+h)-2f(x_0)+f(x_0-h)}{h^2}$$

#### Partial Derivatives: 2D Finite Differences
For a funtion $f\left(x,y\right)\;$ the second order partial derivatives can  be approximated by:
 
$$\frac{\partial^2 f}{\partial x^2}(x_i,y_j) \approx \frac{f(x_i+h_x,y_j)-2f(x_i,y_j)+f(x_i-h_x,y_j)}{h_x^2}$$
 
$$\frac{\partial^2 f}{\partial y^2}(x_i,y_j) \approx \frac{f(x_i,y_j+h_y)-2f(x_i,y_j)+f(x_i,y_j-h_y)}{h_y^2}$$
